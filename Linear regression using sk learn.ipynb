```python
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as lr
from sklearn.metrics import mean_squared_error as mse , r2_score as r2, mean_absolute_error as mae
from math import sqrt as root
```

```python
data = pd.read_csv('sample_data/california_housing_train.csv')
x = data.drop(columns=["population", "longitude", "latitude","median_house_value"], axis=1)
xtr, xte, ytr, yte = tts(x, data["median_house_value"], test_size=0.2, random_state=42)
```

```python
model = lr().fit(xtr, ytr)
ypred = model.predict(xte)
```

```python
r2 = r2(yte, ypred)
rmse = root(mse(yte, ypred))
mae = mae(yte, ypred)
```

```python
print("📊 NumPy Evaluation Metrics")
print("----------------------------")
print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R²   : {r2:.4f}")
```
